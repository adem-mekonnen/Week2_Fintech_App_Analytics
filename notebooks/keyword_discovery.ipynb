```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Load Data
df = pd.read_csv("../data/processed/reviews_cleaned.csv")
stop_words = 'english'

# Run TF-IDF to find unique words for each bank
tfidf = TfidfVectorizer(stop_words=stop_words, max_features=20, ngram_range=(1,2))
tfidf_matrix = tfidf.fit_transform(df['review'].astype(str))
feature_names = tfidf.get_feature_names_out()

print("Top Keywords found in reviews:")
print(feature_names)
```
